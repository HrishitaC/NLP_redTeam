```python
import pandas as pd
import matplotlib.pyplot as plt

withOptions_results_df = pd.read_csv("results/MathQA_withOptions_test_results_revised.csv")
withOptions_score_mean = withOptions_results_df.groupby(["category"])["score_revised"].mean()*100
print(withOptions_score_mean)
ax = withOptions_score_mean.plot(kind='bar', rot=45, figsize=(6, 5), xlabel="category", ylabel="model_accuracy", title="Category wise model accuracy (with options)")
```

```python
noOptions_results_df = pd.read_csv("results/MathQA_noOptions_test_results.csv")
noOptions_score_mean = noOptions_results_df.groupby(["category"])["score"].mean()*100
print(noOptions_score_mean)
ax = noOptions_score_mean.plot(kind='bar', rot = 45, figsize=(6, 5), xlabel="category", ylabel="model_accuracy", title="Category wise model accuracy (no options)")
```

```python
options_df = pd.DataFrame(list(zip(list(withOptions_score_mean.keys()), noOptions_score_mean.values, withOptions_score_mean.values)), columns=["category", "noOptions", "withOptions"])

f = plt.figure()
ax = options_df.plot(x="category", y = ["noOptions", "withOptions"], kind = "bar", rot = 45)
plt.title("Model accuracy without and with options")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
qual_eval = pd.read_csv("results/NLP_redTeam_qualitative.csv")
before_rat = qual_eval.groupby(["category"])["score_beforeRationale"].mean()*100
after_rat = qual_eval.groupby(["category"])["score_afterRationale"].mean()*100
```

```python
print(before_rat)
ax = before_rat.plot(kind='bar', figsize=(6, 5), rot = 45, xlabel="category", ylabel="model_accuracy", title="Category wise model accuracy (beforeRationale)")
```

```python
print(after_rat)
ax = after_rat.plot(kind='bar', figsize=(6, 5), rot = 45,  xlabel="category", ylabel="model_accuracy", title="Category wise model accuracy (afterRationale)")
```

```python
rationale_df = pd.DataFrame(list(zip(list(before_rat.keys()), before_rat.values, after_rat.values)), columns=["category", "before_rationale", "after_rationale"])

f = plt.figure()
ax = rationale_df.plot(x="category", y = ["before_rationale", "after_rationale"], kind = "bar", rot = 45)
plt.title("Model accuracy before and after rationale")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
